```python
# Show the number of object/Categorical features present.
colname=[]
for x in df.columns:
    if df[x].dtype=='object':
        colname.append(x)
colname
```

## <b>Define X (independent variable) and y (Dependent variable) </b>

```python
# Customer name does not impact anything. So we are not considering it.

X=df[['Distributor', 'City', 'Country', 'Latitude',
       'Longitude', 'Channel', 'Sub-channel', 'Product Name', 'Product Class',
       'Quantity', 'Price', 'Month', 'Year', 'Name of Sales Rep',
       'Manager', 'Sales Team']]

y =df['Sales']
```

## ** The dependent variable (Sales) should follow an approximate normal distribution.**

```python
# It seems Slightly Positively skewed.
plt.figure(figsize=(10,6))
sns.distplot(y, hist=True, kde=True)
plt.show()

# with the help of log make it normal distribution
```

```python
y = np.log1p(y)      # Use Log function to make Sales feautures as Normal Distribution.
y.hist();            # Machine understands and predict ND easily.
```

```python
# after use of log funtion our graph looks ND.
plt.figure(figsize=(10,6))
sns.distplot(y, hist=True, kde=True, color='green')
plt.show()

# with the help of log we made it normal distribution
```

## **To check Skewness in 'X'**
### **To check the data is heavyly skewed (right or left) in x variables.**

```python
from scipy.stats import skew
#finding the skewness in each variables
data_num_skew=X.apply(lambda x: skew(x.dropna()))

#subsetting the variables that are highly skewed
data_num_skewed=data_num_skew[(data_num_skew>.75 ) | (data_num_skew<-.75)]

print(data_num_skew)
print(data_num_skewed)

import numpy as np
#apply log +1 transformation for all numeric features with skewness over.75
X[data_num_skewed.index]= np.log1p(X[data_num_skewed.index])
```

```python
# Quantity feature is highly Skewd
data_num_skewed.index   # 4
```

- use of Standard scalar (-3 to 3)

## **VIF-(Varience inflation factor**)

```python
from statsmodels.stats.outliers_influence import  variance_inflation_factor as vif

vif_df= pd.DataFrame()
vif_df['features']=X.columns
vif_df['VIF Factor']= [vif(X.values,i) for i in range(X.shape[1])]
vif_df.round(2)
```

**Will drop 'Latitude' and 'Longitude' column and then again check accuracy overall.**

```python
# See the Vif value of Latitude -- 1039 (Very Big)
X = X.drop(['Latitude'],axis=1)
X = X.drop(['Longitude'],axis=1)
```

## **Split the Data**

```python
from sklearn.model_selection import train_test_split
#split the data into train and test
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=10)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# **Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression
#create a model obj
lm=LinearRegression()
#train the model obj
lm.fit(X_train,y_train)
```

```python
#predict using model
y_pred_new = lm.predict(X_test)
y_pred_new

lm.score(X_train,y_train)   #To check Training accuracy
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np

rmse=np.sqrt(mean_squared_error(y_test,y_pred_new))
print(rmse)
```

- Nothing improve error after standard scalar.
- no need to drop any latitude and longitude bcos it doesnt impact our error.
- Log for Sales --- 0.33

```python
# r2 square means that our model is 96% is accurate on test data.
from sklearn.metrics import r2_score
r_squared = r2_score(y_test, y_pred_new)*100
r_squared
print('r2_squared is :', r_squared)
```

#**Extreme Gradient Boosting (XGBoost) Regressor Model**
- Gradient boosting refers to a class of **ensemble machine learning algorithms** that can be used for classification or regression predictive modeling problems.

- Ensembles are constructed from **decision tree models**.Trees are added one at a time to the ensemble and fit to correct the prediction errors made by prior models. This is a type of ensemble machine learning model referred to as boosting.

- Models are fit using any arbitrary differentiable loss function and gradient descent optimization algorithm. This gives the technique its name, ‚Äúgradient boosting,‚Äù as the loss gradient is minimized as the model is fit, much like a neural network.

```python
from xgboost import XGBRegressor
XGBR = XGBRegressor(n_estimators=1000, max_depth=7)
XGBR.fit(X_train, y_train)

y_pred = XGBR.predict(X_test)
y_pred

XGBR.score(X_train,y_train)   #To check Training accuracy
```

**R2-Square metric**

```python
from sklearn.metrics import r2_score
XGB_R2 = r2_score(y_test, y_pred)*100
XGB_R2
```

**Root mean square error**

```python
rmse=np.sqrt(mean_squared_error(y_test,y_pred))
print(rmse)
```

#**Displaying** **the** **accuracy** **of** **both** **models**

```python
accuracy_df = pd.DataFrame({
    'Model': ['Linear Regression', 'XGBoost'],
    'R2-square': [r_squared, XGB_R2]
})
print(accuracy_df)
```

```python
# XGB model is Best model with 99.99%.
sns.barplot(x=accuracy_df['Model'], y=accuracy_df['R2-square'])
```
